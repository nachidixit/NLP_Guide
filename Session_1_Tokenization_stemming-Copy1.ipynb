```python
from nltk.tokenize import TreebankWordTokenizer, sent_tokenize
token = TreebankWordTokenizer()
```

```python
corpus = """
Hello world! This is a test corpus for tokenizer evaluation.
Tokenizers split text into smaller units called tokens.

Numbers: 123, 45.67, and dates like 2026-06-04.
Special characters: @username, #hashtag, email@example.com.
"""
```

```python
lst_token = token.tokenize(corpus)
lst = set(lst_token)
print("Unique tokens: ", lst)
```

```python
# nltk.download('punkt_tab')
content = sent_tokenize(corpus)
for sent in content:
    print(sent)
```

```python
## Text preprocessing ---> stemming
from nltk.stem import PorterStemmer
port = PorterStemmer()
```

```python
print(lst_token)
```

```python
for i in lst_token:
    new = port.stem(i)
    print(f" {i} -----> {new}")
```

```python
## regexstemmer ---> based on patterns present within word it gives us root words
from nltk.stem import RegexpStemmer
rgx = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
rgx.stem("drinking")
```

```python
rgx.stem("caring")
```

```python
rgx.stem("Winning")
```

```python
## snowball stemmer
from nltk.stem import SnowballStemmer
snow = SnowballStemmer("english")
```

```python
snow.stem("caring")
```

```python
snow.stem("congratulations")
```

```python
snow.stem("winning")
```
