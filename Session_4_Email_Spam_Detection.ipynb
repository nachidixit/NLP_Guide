## Text Preprocessing: - 
1. Handle duplicate ---> Done
2. Handle Mising Values ----> No missing values were present.
3. lowering all the text ---> Done
4. Stopwords ---> Done
5. Lemmatization ---> Done
6. Converting text to vectors
7. ML algorithm

```python
## duplicates
df.duplicated().sum()
```

```python
## removing of duplicates
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
## missing values
df.isnull().sum()
```

```python
## lowercase the text
df["text"] = df["text"].str.lower()
df.head()
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
stop = set(stopwords.words("english"))
def clean_text(text):
    tokens = re.sub(r"\d+", " ", text) ## it will remove all the digits
    tokens = re.sub(r"_"," ", tokens) ## it will remove all the underscores
    tokens = re.sub(r"\W", " ", tokens) ## it will remove all the special characters
    tokens = [i for i in tokens.split() if i not in stop]
    tokens = " ".join(tokens)
    return tokens
df["text"] = df["text"].apply(clean_text)

```

```python
df["text"].iloc[2155]
```

```python
## Lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download("wordnet")
lemma = WordNetLemmatizer()
for i in range(len(df)):
    df["text"].iloc[i] = lemma.lemmatize(df["text"].iloc[i])
    
```

```python
df["text"].iloc[0]
```

## Text to Vectors: - 
1. Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
x = cv.fit_transform(df["text"]).toarray()
y = df["spam"]
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
## classification ---> Baseline
from sklearn.metrics import accuracy_score


model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
y_pred_train = model.predict(x_train)
print(accuracy_score(y_train, y_pred_train))
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(accuracy_score(y_test, y_pred))
```

```python
cv2 = CountVectorizer(max_features=1000, ngram_range=(1,3))
x2 = cv2.fit_transform(df["text"]).toarray()
```
